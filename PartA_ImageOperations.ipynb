#  Ans 2

```python
def imshow(title =" ", image = None, size = 5):
    w,h = image.shape[0],image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
# Resizing
resized = cv2.resize(img, (400, 300))
imshow("Resized image", resized)

# Rotation 90 degree
h,w = img.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((w/2,h/2),90,1)
rotation_img = cv2.warpAffine(img,rotation_matrix,(w,h))
imshow("Rotation", rotation_img)

# Rotation 180 degree
rotation_matrix = cv2.getRotationMatrix2D((w/2,h/2),180,1)                 
rotation_img = cv2.warpAffine(img,rotation_matrix,(w,h))
imshow("Rotation", rotation_img)

# rot_90 = cv2.rotate(img_color, cv2.ROTATE_90_CLOCKWISE)
# rot_180 = cv2.rotate(img_color, cv2.ROTATE_180)

# rotation arbitrary
rotation_matrix = cv2.getRotationMatrix2D((w/2,h/2),45,1)
rotation_img = cv2.warpAffine(img,rotation_matrix,(w,h))
imshow("Rotation", rotation_img)

rotated_img = cv2.transpose(img)
imshow("Transpose", rotated_img)

# Flip image horizontally and vertically
flip_h = cv2.flip(img_color, 1)
imshow("Flip horizontal", flip_h)
flip_v = cv2.flip(img_color, 0)
imshow("Flip vertical", flip_v)
```

# ANS 3

```python
gaussian = cv2.GaussianBlur(img_color, (7,7), 0)
median = cv2.medianBlur(img_color, 7)
bilateral = cv2.bilateralFilter(img_color, 9, 75, 75)

titles = ['Original', 'Gaussian Blur', 'Median Blur', 'Bilateral Filter']
filters = [img_color, gaussian, median, bilateral]

# comparison
plt.figure(figsize=(12,6))
for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(cv2.cvtColor(filters[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])

plt.show()
```

```python
# 4. Adjust brightness and contrast manually

a = 1.4  # Contrast
b = 40   # Brightness
adjusted = cv2.convertScaleAbs(img_color, alpha=a, beta=b)
imshow("Brightness and Contrast adjusted", adjusted)
```

# Ans 5


```python
# Edge Detection
#  Canny
gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(gray,100, 200)

# Sobel and Laplacian 
sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)
sobel_combined = cv2.convertScaleAbs(sobelx + sobely)
laplacian = cv2.Laplacian(gray, cv2.CV_64F)
laplacian_abs = cv2.convertScaleAbs(laplacian)

plt.figure(figsize=(12,8))
images = [gray, canny, sobel_combined, laplacian_abs]
titles = ['Original Grayscale', 'Canny Edge', 'Sobel Edge', 'Laplacian Edge']

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis("off")
plt.show()
```
